```python
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df
```

GENDER DISTRIBUTION VISUALIZATION USING COUNT PLOT


```python
plt.figure(figsize = (4,5))
ax = sns.countplot(data= df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean',"ReadingScore":"mean","WritingScore":"mean"})
gb
```

```python
sns.heatmap(gb,annot= True)# to know the values inside heat map we use annot keyword
plt.title("Relationship between Parent's Education and Student Score")
plt.show()
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean',"ReadingScore":"mean","WritingScore":"mean"})
gb1
```

```python
sns.heatmap(gb1,annot= True)# to know the values inside heat map we use annot keyword
plt.title("Relationship between Parent's Marital Status  and Student's Score")
plt.show()
```

DISTRIBUTION OF ETHINIC GROUPS

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
l = ["group A","groupB","groupC","groupD","groupE"]
mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mylist,labels = l,autopct = "%1.2f%%")
plt.title("Distibution of EthnicGroups")
plt.show()
```
